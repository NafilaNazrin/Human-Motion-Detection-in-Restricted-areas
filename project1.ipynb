```python
import cv2
import smtplib
from email.mime.text import MIMEText
import numpy as np
import time

# Client email setup
CLIENT_EMAIL = 'test.client@domain.com'  # client email
YOUR_EMAIL = 'example@gmail.com' 
YOUR_PASSWORD = 'ytmm bhwo xhlc xvcg'  # app password here

# Cooldown configuration
COOLDOWN_TIME = 10  # Cooldown time in seconds
WARM_UP_TIME = 10   # Warm-up time to stabilize the background subtractor
last_alert_time = 0  # Last time an alert was sent

def send_email_alert():
    msg = MIMEText("ðŸš¨ ALERT: Human detected in restricted area!")
    msg['Subject'] = 'Motion Detection Alert'
    msg['From'] = YOUR_EMAIL
    msg['To'] = CLIENT_EMAIL
    
    with smtplib.SMTP('smtp.gmail.com', 587) as server:
        server.starttls()
        server.login(YOUR_EMAIL, YOUR_PASSWORD)
        server.sendmail(msg['From'], msg['To'], msg.as_string())
        print(f"Email sent to {CLIENT_EMAIL}!")

def motion_detection():
    global last_alert_time  
    cap = cv2.VideoCapture(0)
    fgbg = cv2.createBackgroundSubtractorMOG2()
    
    # Warm-up phase: Capture frames but do not send alerts during this time
    print("Warming up the background subtractor...")
    start_time = time.time()
    while time.time() - start_time < WARM_UP_TIME:
        ret, frame = cap.read()
        if not ret:
            break
        fgbg.apply(frame)  # Let the background stabilizer process frames
        cv2.imshow('Motion Detection (Warm-up)', frame)
        if cv2.waitKey(30) & 0xFF == 27:  # Press 'ESC' to exit
            cap.release()
            cv2.destroyAllWindows()
            return 
    
    print("Warm-up complete. Motion detection is now active.")
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        fgmask = fgbg.apply(frame)
        contours, _ = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        motion_detected = False

        for contour in contours:
            if cv2.contourArea(contour) > 10000:  # Area threshold 
                motion_detected = True
                (x, y, w, h) = cv2.boundingRect(contour)
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Check if motion is detected and if cooldown has passed
        current_time = time.time()
        if motion_detected and (current_time - last_alert_time > COOLDOWN_TIME):
            send_email_alert()
            last_alert_time = current_time  # Update last alert time

        cv2.imshow('Motion Detection', frame)
        cv2.imshow('FG Mask', fgmask)

        if cv2.waitKey(30) & 0xFF == 27:  # Press 'ESC' to exit
            break

    cap.release()
    cv2.destroyAllWindows()
    
motion_detection()
```
